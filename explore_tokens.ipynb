```python
import pandas as pd
import matplotlib.pyplot as plt

input_df = pd.read_csv('output/token_count/text_sql_tokens_count_gpt-4o.csv')
```

```python
# Calculate cumulative values
# The first iteration
input_df['num_tokens_cumulative'] = input_df['num_tokens'].cumsum()

```

```python
# Total number of tokens
print('Total number of tokens in GPT-4o with 10 examples for the current table:', input_df['num_tokens_cumulative'].iloc[-1])
print(f"Total cost per run: {input_df['num_tokens_cumulative'].iloc[-1] * 5 / 1000000}")
```

```python
plt.figure(figsize=(14, 8))

# Plot for 'gold_prompt_tokens_cumulative'
plt.subplot(3, 1, 3)
plt.plot(input_df['num_tokens_cumulative'], label='GPT-4o 10 examples')
plt.title('Cumulative Prompt Tokens')
plt.xlabel('Index')
plt.ylabel('Cumulative Tokens')
plt.legend()
plt.tight_layout()
# plt.savefig(f'plots/cumulative_tokens_count.png')
plt.show()
```
