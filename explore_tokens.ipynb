```python
import pandas as pd
import matplotlib.pyplot as plt

EXPERIMENT_NAME_1 = 'tokens_count'
gpt_4o = pd.read_csv(EXPERIMENT_NAME_1 + '_gpt-4o.csv')
gpt_3_5_turbo = pd.read_csv(EXPERIMENT_NAME_1 + '_gpt-3.5-turbo.csv')

EXPERIMENT_NAME_2 = '3_10ex_tokens_count'
gpt_4o_3_10ex = pd.read_csv(EXPERIMENT_NAME_2 + '_gpt-4o.csv')
gpt_3_5_turbo_3_10ex = pd.read_csv(EXPERIMENT_NAME_2 + '_gpt-3.5-turbo.csv')

EXPERIMENT_NAME_3 = '10ex_tokens_count'
gpt_4o_10ex = pd.read_csv(EXPERIMENT_NAME_3 + '_gpt-4o.csv')
gpt_3_5_turbo_10ex = pd.read_csv(EXPERIMENT_NAME_3 + '_gpt-3.5-turbo.csv')
```

```python
# Calculate cumulative values
# The first iteration
gpt_4o['gold_prompt_chars_cumulative'] = gpt_4o['gold_prompt_chars'].cumsum()
gpt_4o['gold_prompt_words_cumulative'] = gpt_4o['gold_prompt_words'].cumsum()
gpt_4o['gold_prompt_tokens_cumulative'] = gpt_4o['gold_prompt_tokens'].cumsum()

gpt_3_5_turbo['gold_prompt_chars_cumulative'] = gpt_3_5_turbo['gold_prompt_chars'].cumsum()
gpt_3_5_turbo['gold_prompt_words_cumulative'] = gpt_3_5_turbo['gold_prompt_words'].cumsum()
gpt_3_5_turbo['gold_prompt_tokens_cumulative'] = gpt_3_5_turbo['gold_prompt_tokens'].cumsum()

# The second iteration
gpt_4o_3_10ex['gold_prompt_chars_cumulative'] = gpt_4o_3_10ex['gold_prompt_chars'].cumsum()
gpt_4o_3_10ex['gold_prompt_words_cumulative'] = gpt_4o_3_10ex['gold_prompt_words'].cumsum()
gpt_4o_3_10ex['gold_prompt_tokens_cumulative'] = gpt_4o_3_10ex['gold_prompt_tokens'].cumsum()

gpt_3_5_turbo_3_10ex['gold_prompt_chars_cumulative'] = gpt_3_5_turbo_3_10ex['gold_prompt_chars'].cumsum()
gpt_3_5_turbo_3_10ex['gold_prompt_words_cumulative'] = gpt_3_5_turbo_3_10ex['gold_prompt_words'].cumsum()
gpt_3_5_turbo_3_10ex['gold_prompt_tokens_cumulative'] = gpt_3_5_turbo_3_10ex['gold_prompt_tokens'].cumsum()

# The third iteration
gpt_4o_10ex['gold_prompt_chars_cumulative'] = gpt_4o_10ex['gold_prompt_chars'].cumsum()
gpt_4o_10ex['gold_prompt_words_cumulative'] = gpt_4o_10ex['gold_prompt_words'].cumsum()
gpt_4o_10ex['gold_prompt_tokens_cumulative'] = gpt_4o_10ex['gold_prompt_tokens'].cumsum()

gpt_3_5_turbo_10ex['gold_prompt_chars_cumulative'] = gpt_3_5_turbo_10ex['gold_prompt_chars'].cumsum()
gpt_3_5_turbo_10ex['gold_prompt_words_cumulative'] = gpt_3_5_turbo_10ex['gold_prompt_words'].cumsum()
gpt_3_5_turbo_10ex['gold_prompt_tokens_cumulative'] = gpt_3_5_turbo_10ex['gold_prompt_tokens'].cumsum()
```

```python
# Total number of tokens
print('Total number of tokens in GPT-4o with 3 examples for every table, 100 for the current table and 20 unique for the current column:', gpt_4o['gold_prompt_tokens_cumulative'].iloc[-1])
print('Total number of tokens in GPT-3.5-turbo with 3 examples for every table, 100 for the current table and 20 unique for the current column:', gpt_3_5_turbo['gold_prompt_tokens_cumulative'].iloc[-1])
print('Total number of tokens in GPT-4o with 3 examples for every table, 10 for the current table:', gpt_4o_3_10ex['gold_prompt_tokens_cumulative'].iloc[-1])
print('Total number of tokens in GPT-3.5-turbo with 3 examples for every table, 10 for the current table:', gpt_3_5_turbo_3_10ex['gold_prompt_tokens_cumulative'].iloc[-1])
print('Total number of tokens in GPT-4o with 10 examples for the current table:', gpt_4o_10ex['gold_prompt_tokens_cumulative'].iloc[-1])
print('Total number of tokens in GPT-3.5-turbo with 10 examples for the current table:', gpt_3_5_turbo_10ex['gold_prompt_tokens_cumulative'].iloc[-1])
```

```python
plt.figure(figsize=(14, 8))

# Plot for 'gold_prompt_chars_cumulative'
plt.subplot(3, 1, 1)
plt.plot(gpt_4o['gold_prompt_chars_cumulative'], label='GPT-4.0')
plt.plot(gpt_3_5_turbo['gold_prompt_chars_cumulative'], label='GPT-3.5 Turbo')
plt.plot(gpt_4o_3_10ex['gold_prompt_chars_cumulative'], label='GPT-4.0 3+10 examples')
plt.plot(gpt_3_5_turbo_3_10ex['gold_prompt_chars_cumulative'], label='GPT-3.5 Turbo 3+10 examples')
plt.plot(gpt_4o_10ex['gold_prompt_chars_cumulative'], label='GPT-4.0 10 examples')
plt.plot(gpt_3_5_turbo_10ex['gold_prompt_chars_cumulative'], label='GPT-3.5 Turbo 10 examples')
plt.title('Cumulative Gold Prompt Characters')
plt.xlabel('Index')
plt.ylabel('Cumulative Characters')
plt.legend()

# Plot for 'gold_prompt_words_cumulative'
plt.subplot(3, 1, 2)
plt.plot(gpt_4o['gold_prompt_words_cumulative'], label='GPT-4.0')
plt.plot(gpt_3_5_turbo['gold_prompt_words_cumulative'], label='GPT-3.5 Turbo')
plt.plot(gpt_4o_3_10ex['gold_prompt_words_cumulative'], label='GPT-4.0 3+10 examples')
plt.plot(gpt_3_5_turbo_3_10ex['gold_prompt_words_cumulative'], label='GPT-3.5 Turbo 3+10 examples')
plt.plot(gpt_4o_10ex['gold_prompt_words_cumulative'], label='GPT-4.0 10 examples')
plt.plot(gpt_3_5_turbo_10ex['gold_prompt_words_cumulative'], label='GPT-3.5 Turbo 10 examples')
plt.title('Cumulative Gold Prompt Words')
plt.xlabel('Index')
plt.ylabel('Cumulative Words')
plt.legend()

# Plot for 'gold_prompt_tokens_cumulative'
plt.subplot(3, 1, 3)
plt.plot(gpt_4o['gold_prompt_tokens_cumulative'], label='GPT-4.0')
plt.plot(gpt_3_5_turbo['gold_prompt_tokens_cumulative'], label='GPT-3.5 Turbo')
plt.plot(gpt_4o_3_10ex['gold_prompt_tokens_cumulative'], label='GPT-4.0 3+10 examples')
plt.plot(gpt_3_5_turbo_3_10ex['gold_prompt_tokens_cumulative'], label='GPT-3.5 Turbo 3+10 examples')
plt.plot(gpt_4o_10ex['gold_prompt_tokens_cumulative'], label='GPT-4.0 10 examples')
plt.plot(gpt_3_5_turbo_10ex['gold_prompt_tokens_cumulative'], label='GPT-3.5 Turbo 10 examples')
plt.title('Cumulative Gold Prompt Tokens')
plt.xlabel('Index')
plt.ylabel('Cumulative Tokens')
plt.legend()

plt.tight_layout()
plt.savefig(f'plots/cumulative_tokens_count.png')
plt.show()
```

```python
# Plot the tokens for different databases
aggregated_data_gpt_4o = gpt_4o.groupby('database_name')['gold_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_3_5_turbo = gpt_3_5_turbo.groupby('database_name')['gold_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_4o_3_10ex = gpt_4o_3_10ex.groupby('database_name')['gold_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_3_5_turbo_3_10ex = gpt_3_5_turbo_3_10ex.groupby('database_name')['gold_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_4o_10ex = gpt_4o_10ex.groupby('database_name')['gold_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_3_5_turbo_10ex = gpt_3_5_turbo_10ex.groupby('database_name')['gold_prompt_tokens'].sum().reset_index()

# Plot the aggregated data
plt.figure(figsize=(10, 6))
plt.bar(aggregated_data_gpt_4o['database_name'], aggregated_data_gpt_4o['gold_prompt_tokens'], label='GPT-4.0')
plt.bar(aggregated_data_gpt_4o_3_10ex['database_name'], aggregated_data_gpt_4o_3_10ex['gold_prompt_tokens'], label='GPT-4.0 3+10 examples')
plt.bar(aggregated_data_gpt_4o_10ex['database_name'], aggregated_data_gpt_4o_10ex['gold_prompt_tokens'], label='GPT-4.0 10 examples')
plt.title('Total Number of Tokens for Each Database for the Gold Prompts')
plt.xlabel('Database Name')
plt.ylabel('Total Gold Prompt Tokens')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.savefig(f'plots/total_tokens_count_gold.png')
plt.show()
```

```python
aggregated_data_gpt_4o_pred = gpt_4o.groupby('database_name')['pred_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_3_5_turbo_pred = gpt_3_5_turbo.groupby('database_name')['pred_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_4o_3_10ex_pred = gpt_4o_3_10ex.groupby('database_name')['pred_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_3_5_turbo_3_10ex_pred = gpt_3_5_turbo_3_10ex.groupby('database_name')['pred_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_4o_10ex_pred = gpt_4o_10ex.groupby('database_name')['pred_prompt_tokens'].sum().reset_index()
aggregated_data_gpt_3_5_turbo_10ex_pred = gpt_3_5_turbo_10ex.groupby('database_name')['pred_prompt_tokens'].sum().reset_index()

# Plot the aggregated data
plt.figure(figsize=(10, 6))
plt.bar(aggregated_data_gpt_4o_pred['database_name'], aggregated_data_gpt_4o_pred['pred_prompt_tokens'], label='GPT-4o')
plt.bar(aggregated_data_gpt_4o_3_10ex_pred['database_name'], aggregated_data_gpt_4o_3_10ex_pred['pred_prompt_tokens'], label='GPT-4o 3+10 examples')
plt.bar(aggregated_data_gpt_4o_10ex_pred['database_name'], aggregated_data_gpt_4o_10ex_pred['pred_prompt_tokens'], label='GPT-4o 10 examples')
plt.title('Total Number of Tokens for Each Database for the Prediction Prompts')
plt.xlabel('Database Name')
plt.ylabel('Total Prediction Prompt Tokens')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f'plots/total_tokens_count_pred.png')
plt.show()
```
